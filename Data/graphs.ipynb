```python

# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

# File to Load 
mentorship_data = "mentorship_eligibilty.csv"
retiring_data = "retiring_titles.csv"
unique_data = "unique_titles.csv"

# Read the Data
mentorship_df = pd.read_csv(mentorship_data)
retiring_df = pd.read_csv(retiring_data)
unique_df = pd.read_csv(unique_data)
```

```python
titles_total = unique_df.groupby(["title"]).count()["emp_no"]
print(titles_total)
```

```python
mentorship_total = mentorship_df.groupby(["title"]).count()["emp_no"]
print(mentorship_total)
```

```python
data_summary_df = pd.DataFrame(
    {"Total Retiring" : titles_total,
    "Total Mentors" : mentorship_total})
data_summary_df
```

```python
plt.pie(titles_total, labels= ["Assistant Engineer", "Engineer", "Manager", "Senior Engineer", "Senior Staff", "Staff", "Technique Leader"], 
autopct='%1.2f%%', radius=2)
plt.savefig('total_retiring_pie.png')
plt.show()
```

```python
titles_per_mentor = titles_total/mentorship_total
titles_per_mentor
```

```python
titles = ("Assistant Engineer", "Engineer", "Manager", "Senior Engineer", "Senior Staff", "Staff", "Technique Leader")
y_pos = np.arange(len(titles))
plt.barh(titles, titles_per_mentor)
plt.xlabel("New Openings Per Mentor")
```
